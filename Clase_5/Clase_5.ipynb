* #### Por qué analizar las vibraciones?
  + Para poder **suprimir** vibraciones y así, por ejemplo:
    * **Proporcionar seguridad**: respuestas de edificios frente a terremotos, de puentes frente a vientos caóticos.
    + **Contribuir a la comodidad**: sistemas de suspensión de vehículos, disminución de ruido en cabinas.
    * **Proteger equipos**: instrumental médico de alta precisión (robot DaVinci), instrumentos astronómicos como telescopios, interferómetros.
    * **Proteger a los seres humanos**: la exposición a vibración puede producir visión borrosa, pérdida de equilibrio, pérdida de concentración, problemas de circulación de sangre en los tejidos (dedos blancos).
    * **Evitar pérdidas económicas**.

* #### Cómo se puede lograr?
  + Por ejemplo, mediante el uso de:
    * **Amortiguamiento de las vibraciones**, reduciendo peaks en regiones de resonancia.
    * **Aislamiento de las vibraciones**, reduciendo la transmisibilidad de fuerzas responsables de accionar a los sistemas.

* #### Parámetros discretos de un sistema mecánico que vibra
  + En un sistema se distinguen tres parámetros discretos básicos:
    1. Su ***masa***, representada por un elemento de masa rígida. 
      + Este es capaz de almacenar energía cinética. 
    2. Su ***amortiguamiento***, representado por un amortiguador.
      + Este representa la pérdida gradual de energía del sistema.
    3. Su ***elasticidad***, representado por un resorte.
      + Este parámetro almacena energía potencial.
  + Pensar de manera discreta sobre estos tres parámetros, significa suponer que el sistema es una **combinación** de masas interconectadas por resortes flexibles y elementos amortiguadores.

* #### Resortes
  + Los resortes son quizás el parámetro discreto más importante en análisis de vibraciones. La razón se debe a que este parámetro es el link que une a dos partículas del cuerpo en vibración.
  + El resorte es un medio continuo que tiene su propia masa, pero que en general, si se compara con la masa del sistema vibratorio es tan pequeña que se suele despreciar. 
  + El largo del resorte cuando no está accionado por ninguna fuerza externa se llama ***largo no estirado***. Dado que el resorte es de material flexible, cualquier variación en su largo, en un ${\text{x}}$ dado (${\text{x}}$ es una variable continua, positiva o negativa), debido a la acción de una fuerza $\text{F}$ aplicada al resorte, será una función de ${\text{x}}$:

\begin{align*}
{\text{F}}&={\text{f(x)}}
\end{align*}

* La forma apropiada de ${\text{f(x)}}$ queda determinada por el material del resorte. Si ${\text{f(x)}}$ es infinitamente diferenciable en ${\text{x=0}}$, entonces por serie de Taylor en torno a ${\text{x=0}}$ se usa la expansión (conocida como *MacLaurin expansion*):

\begin{align*}
{\text{F}}&={\text{k}}_{0}+{\text{k}}_{1}{\text{x}}^{}+{\text{k}}_{1}{\text{x}}^{2}+\ldots
\end{align*}

* Ya que ${\text{x}}$ representa la variación en el largo del resorte (se suele llamar también **extensión** y puede ser positiva o negativa), desde su largo no estirado, si ${\text{x=0}}$, entonces ${\text{F=0}}$. Así que, ${\text{k}}_{0}=0$.

* Además, por convención, se suele considerar una ecuación clásica no lineal, de la forma:

\begin{align*}
{\text{F}}&=-{\text{k}}_{1}{\text{x}}^{}-{\text{k}}_{3}{\text{x}}^{3}-{\text{k}}_{5}{\text{x}}^{5}-\ldots
\end{align*} 

donde el signo menos se usa para asegurar que la fuerza siempre actúa, ya sea para acortar el resorte si éste está estirado (extensión positiva), o bien, para alargar el resorte si éste está comprimido (extensión negativa).

* En realidad, todos los resortes son **no lineales.** Sin embargo, hay muchos casos, como en esta asignatura, donde la extensión ${\text{x}}$ se puede asumir suficientemente pequeña para que los términos no-lineales de la ecuación anterior, sean despreciables en comparación con el primer término que es:

* ${\text{F}}\approx-{\text{k}}_{1}{\text{x}}^{}$

 * Por lo tanto, un resorte **lineal** obedece a una **ley fuerza vs desplazamiento** de la forma:

\begin{align}
{\text{F}}&=-{\text{k}}{\text{x}}^{}
\label{eq1}\tag{1}
\end{align}

* donde ${\text{k}}$, que es siempre positiva, se llama la rigidez del resorte o constante del resorte. Tiene dimensiones de fuerza por unidad de longitud.

* #### Sistema vibratorio lineal
    * El sistema de la Fig. 5, forma una relación (mapeo, función) entre la excitación y la respuesta. Si se escribe como función del tiempo:
      + ${\text{x(t)}}={\text{L}}\big\{{\text{f(t)}}\big\}$
    * **Linealidad** quiere decir que podríamos separar una excitación en partes simples y usar superposición en la respuesta para recuperar la salida global del sistema:
      + ${\text{L}}\big\{{\text{a}}\cdot {\text{f}}_{1}{\text{(t)}}+{\text{b}}\cdot {\text{f}}_{2}{\text{(t)}} \big\}={\text{a}}\cdot{\text{L}}\big\{{\text{f}}_{1}{\text{(t)}}\big\}+{\text{b}}\cdot{\text{L}}\big\{{\text{f}}_{2}{\text{(t)}} \big\}$

* #### Sistema vibratorio invariante en el tiempo
    * **Invariante en el tiempo** significa que si la salida del sistema es:
        + ${\text{x}_{1}}{\text{(t)}}={\text{L}}\big\{{\text{f(t)}}\big\}$
    * entonces, un corrimiento en el tiempo de la excitación de entrada, crea un corrimiento igual en el tiempo en la respuesta del sistema: 
        + ${\text{x}_{2}}{\text{(t)}}={\text{L}}\big\{{\text{f(t-}}{\text{t}}_{o})\big\}={\text{x}_{1}}{\text{(t-}}{\text{t}}_{o})$

* #### Ecuación de movimiento del sistema lineal de un grado de libertad
    + ${\text{m}}\,{\ddot{{\text{x}}}}+{\text{c}}\,{\dot{{\text{x}}}}+{\text{k}}\,{\text{x}}={\text{f(t)}}$

#### Vibraciones libres no amortiguadas ($f(t)=0$ y $c=0$)

La Fig. 7, muestra tres casos de sistemas de un grado de libertad para vibraciones libres no amortiguadas: pédulo simple, masa-resorte y disco-barra:

<p style="text-align:center">
	<img src="ejemplos.png" alt="ejemplos 1DOF" width=45%/></a><br>

Para el caso del sistema masa-resorte, se obtiene la ecuación con sus dos condiciones iniciales:

\begin{align}
m\ddot{x}+kx=0; \quad x(0)=x_{0}, \quad \dot{x}(0)=\dot{x}_{0}
\label{eq2}\tag{2}
\end{align}

que es la ecuación clásica de movimiento de un **oscilador armónico**. Esta forma de ecuación de movimiento es muy importante en análisis de vibraciones. 

* La solución de esta ecuación de movimiento es:

\begin{align*}
x(t)&=c_{1}e^{r_{1}t}+c_{2}e^{r_{2}t}
\end{align*}

* Los valores de $r_{1}$ y $r_{2}$ se encuentran por la ecuación característica:

\begin{align*}
mr^{2}+k=0 \quad \rightarrow \quad r_{1,2} = \pm \,j\sqrt{\frac{k}{m}}=\pm \, j\,\omega_{0}
\end{align*}

* La solución general será la suma:

\begin{align*}
x(t)&=c_{1}e^{\,j\,\omega_{0}t}+c_{2}e^{\,-j\,\omega_{0}t}
\end{align*}

* Por conveniencia matemática, suponer que

\begin{align*}
c_{2}&=c_{1}^{*}
\end{align*}

Sea $c_{1}= \frac{c}{2}e^{-j\phi}$, tal que

\begin{align*}
c_{2}&=c_{1}^{*}=\frac{c}{2}e^{j\phi}
\end{align*}

donde $c$ y $\phi$ son constantes. De modo que:

\begin{align*}
x(t)&=c\,{\text{cos}}\,(\omega_{0}t-\phi)
\end{align*}

<p style="text-align:center">
	<img src="respuesta.png" alt="respuesta" width=35%/></a><br>
    <strong> Figura 8: Respuesta $x(t)$ de acuerdo a condiciones iniciales</strong>
</p>

Para determinar $c$ y $\phi$, se usan las condiciones iniciales:

\begin{align*}
x(0)&=x_{0}=c\,{\text{cos}}\phi
\end{align*}

\begin{align*}
\dot{x}(t)&=-c\,\omega_{0}\,{\text{sen}}\,(\omega_{0}\,t-\phi)
\end{align*}

\begin{align*}
\dot{x}(0)&=\dot{x}_{0}=c\,\omega_{0}\,{\text{sen}}\phi
\end{align*}

de donde $c$ es igual a:

\begin{align*}
c&=\sqrt{x_{0}^{2}+\bigg( \frac{\dot{x}_{0}}{\omega_{0}}\bigg)^{2}}
\end{align*}

y además,

\begin{align*}
tg(\phi)&=\frac{{\text{sen}}(\phi)}{{\text{cos}}(\phi)}=\frac{\dot{x}_{0}}{x_{0}\omega_{0}}
\end{align*}

de donde:

\begin{align*}
\phi &= tg^{-1}\bigg(\frac{\dot{x}_{0}}{x_{0}\omega_{0}}\bigg)
\end{align*}

En estas ecuaciones, 
* $c$ se llama **valor peak** (m) y representa la amplitud del desplazamiento.
* $T$ es el **período de la vibración** (s) en torno al punto de equilibrio.
* $f_{0}=1/T$ es la **frecuencia natural** de la vibración (Hz).
* $\omega_{0}=2\pi f_{0}$ es la **frecuencia natural** de la vibración (rad/s).
* $\phi$ es el **ángulo de fase** (rad).
* $v=\dot{x},$ es la **velocidad** (m/s).
* $a=\ddot{x},$ es la **aceleración** (m/$s^{2}$).
* $c_{pp}=2c,$ es el **valor peak to peak** de la vibración.
* $c_{RMS}=0.707\cdot c,$ es el **valor RMS** de la vibración.

#### Frecuencia natural de vibración

* La frecuencia $\omega_{0}$ con la cual vibra libremente el sistema sin amortiguamiento, se llama su frecuencia natural de vibración:

\begin{align*}
\omega_{0}&=\sqrt{\frac{k}{m}}
\end{align*}

* Solamente depende de las características del sistema (Fig. 7).
* Si aumenta la masa del sistema vibratorio, entonces disminuirá $\omega_{0}$.
* Si aumenta la rigidez del sistema, entonces aumentará $\omega_{0}$.

### Ejemplos de comportamiento tipo resorte equivalente: Rigidez equivalente

* Al menos hay tres tipos de comportamiento **tipo resorte** surge en diversas situaciones de vibración, estas son:
    1. **Longitudinal** (vibración axial o a lo largo), 
    2. **Transversal** (vibración en dirección perpendicular al largo)
    3. **Torsional** (vibración rotacional a lo largo).

#### 1. Resorte equivalente de una barra en vibración longitudinal

* Una barra delgada de largo $L$, (con masa despreciable en comparación con la masa de un sistema), fija en $x=0$ y con una fuerza axial $F$ en $x=L,$ puede ser considerada como un resorte, con rigidez equivalente $k_{eq}$ longitudinal dada por:

\begin{align*}
k_{eq}&=\frac{E\,A}{L}
\label{eq3}\tag{3}
\end{align*}

siendo $E$ el módulo de Young también llamado módulo de elasticidad y $A(x)$ la sección de área de la barra.

<p style="text-align:center">
	<img src="axial.png" alt="resorte equivalente" width=35%/></a><br>
    <strong> Figura 9: Barra en vibración longitudinal considerada como resorte.</strong>
</p>

* #### Desarrollo

+ En efecto, supongamos que la rigidez equivalente, $k_{eq}$, de la barra sujeta a vibraciones a su largo, se puede expresar brevemente como:

\begin{align}
k_{eq}&=\frac{F}{\delta}
\label{eq4}\tag{4}
\end{align}

donde $\delta$ es un desplazamiento axial (longitudinal) de la punta de la barra ($x=L$).


+ Como se ve en la Fig. 9, llamemos $u(x)$ al desplazamiento axial de un punto de la barra, a una distancia $x$ del extremo $x=0.$ El desplazamiento $u(x)$ satisface la siguiente ecuación diferencial:

\begin{align}
E\,A(x)\,\frac{du}{dx}&=F(x),\quad {\text{para}}\,\,0<x<L
\label{eq5}\tag{5}
\end{align}

donde $E\,A(x)$ se llama la **rigidez axial** de la barra y $F(x)$ es la fuerza axial, en este caso, la misma para cualquier punto $x$, e igual a la fuerza $F$ aplicada en $x=L$, es decir, $F(x)=F.$ 

+ Ya que la barra está fija en $x=0$, entonces

\begin{align*}
u(0)&=0
\end{align*}

Integrando la Ec. ([5](#mjx-eqn-eq5)), se obtiene:

\begin{align*}
u(x)&=F\,\int_{0}^{x}\frac{1}{E\,A(\zeta)}d\zeta
\end{align*}

Pero, en $x=L$, el desplazamiento de la punta de la barra debe ser igual a $\delta$, de modo que

\begin{align*}
u(L) &=\delta = F\,\int_{0}^{L}\frac{1}{E\,A(\zeta)}d\zeta
\end{align*}

Por lo tanto, reemplazando en la Ec. ([4](#mjx-eqn-eq1)), 

\begin{align*}
k_{eq}&=\frac{F}{\delta}=\bigg[\, \int_{0}^{L}\frac{1}{E\,A(\zeta)}d\zeta \,\bigg]^{-1}
\end{align*}

Si suponemos que la barra es uniforme, entonces la rigidez axial será constante, esto es, $E\,A(\zeta)=E\,A={\text{constante}}$, 

\begin{align*}
k_{eq}&=\bigg[\, \frac{1}{E\,A}\int_{0}^{L}d\zeta \,\bigg]^{-1}=\bigg[\, \frac{L}{E\,A}\,\bigg]^{-1}
\label{eq6}\tag{6}
\end{align*}

Por lo tanto, se obtiene el mismo resultado de la Ec. ([3](#mjx-eqn-eq3)):

\begin{align*}
k_{eq}&=\frac{E\,A}{L}
\end{align*}

Para algunos materiales,

| Material | Módulo de Young E (N/m^2)| Densidad (kg/m^3) | Módulo de corte G (N/m^2) |
| --- | --- | --- |--- |
| Acero | 2x10^11 | 7.8x10^3 |8x10^10 |
| --- | --- | --- | --- |
| Madera | 5.4x10^9 | 6x10^2|--- | 
| --- | --- | --- |--- |
| Concreto | 3,8x10^9 | 1.3x10^2 |--- |

* Otro caso de interés de rigidez axial equivalente, es aquel que surge cuando la configuración es una barra seccionalmente uniforme, como se muestra en la Fig. 7. 

<p style="text-align:center">
	<img src="seccional.png" alt="barra seccional" width=35%/></a><br>

En esta situación y siguiendo el resultado de la Ec. ([6](#mjx-eqn-eq6)), la rigidez axial equivalente será:

\begin{align*}
k_{eq}&=\bigg[\, \frac{L_{1}}{E\,A_{1}}+\frac{L_{2}}{E\,A_{2}}\,\bigg]^{-1}
\label{eq7}\tag{7}
\end{align*}

#### 2. Resorte equivalente de una barra en vibración en dirección perpendicular a su largo

* En la Fig. 11, $L$, representa el largo de una barra delgada; $E$ es su modulo de Young e $I(x)$ es el momento de inercia de la sección transversal a una distancia $x$ del extremo $x=0$. La rigidez equivalente $k_{eq}$ transversal de la barra es:

\begin{align*}
k_{eq}&=\frac{3EI}{L^{3}}
\label{eq8}\tag{8}
\end{align*}

<p style="text-align:center">
	<img src="perpendicular.png" alt="resorte equivalente perpendicular" width=35%/></a><br>
    <strong> Figura 11: Barra en vibración perpendicular a su largo considerada como resorte.</strong>
</p>

* #### Desarrollo

+ En efecto, como muestra la Fig. 11, supongamos que la barra está fija en $x=0$ y una fuerza transversal $F$ la acciona en $x=L$. Llamemos $w(x)$ al desplazamiento transversal. En este caso, la **rigidez de flexión** de la barra será: $E\,I(x)$, donde $I(x)$ es el momento de inercia de la sección de área transversal. El desplazamiento transversal $w(x)$, satisface la ecuación diferencial:

\begin{align}
E\,I(x)\,\frac{d^{2}w(x)}{dx^{2}}=M(x),\quad {\text{para}}\,\,0<x<L
\label{eq9}\tag{9}
\end{align}

donde $M(x)$ es el momento de flexión de la fuerza, en este caso, $F\cdot (L-x)$. 

+ Como se ve en la figura, en $x=0$, tanto el desplazamiento transversal como también, la curva de desplazamiento, deben ser cero. De este modo, las condiciones de borde del problema son:

\begin{align*}
w=0,\frac{dw}{dx}=0,\quad {\text{en}}\,\, x=0.
\end{align*}

Integrando dos veces la Ec. ([9](#mjx-eqn-eq9)), y considerando las condiciones de borde, se obtiene la solución para el desplazamiento transversal

\begin{align}
w(x)&=F\,\int_{0}^{x} \bigg( \int_{0}^{\zeta} \frac{L-\psi}{E\,I(\psi)}d\psi\bigg) d\zeta
\label{eq10}\tag{10}
\end{align}

* Haciendo $w(L)=\delta$ y usando el mismo método analítico del caso anterior del resorte equivalente de una barra en vibración longitudinal (axial), se obtiene el resorte equivalente de la barra en vibración perpendicular a su largo: 

\begin{align}
k_{eq}&=\frac{F}{\delta}=\frac{F}{w(L)}=\bigg[\, \int_{0}^{L} \bigg( \int_{0}^{\zeta} \frac{L-\psi}{E\,I(\psi)}d\psi\bigg) d\zeta\,\bigg]^{-1}
\label{eq11}\tag{11}
\end{align}

Si la barra es uniforme, la rigidez de flexión será: $E\,I(x)=E\,I={\text{constante}}$, de modo que la Ec. ([11](#mjx-eqn-eq11)) se reduce a:

\begin{align*}
k_{eq}&=\bigg[\, \frac{1}{E\,I}\int_{0}^{L} \bigg( \int_{0}^{\zeta} (L-\psi)\,d\psi\bigg) d\zeta\,\bigg]^{-1}\\
&=\bigg[\, \frac{1}{E\,I}\int_{0}^{L} \big( L\zeta-\frac{1}{2}\zeta^{2}\big) d\zeta\,\bigg]^{-1}\\
&=\frac{3E\,I}{L^{3}}
\end{align*}

Por lo tanto, se obtiene el mismo resultado de la Ec. ([8](#mjx-eqn-eq8)).

#### 3. Resorte equivalente de una barra en vibración torsional

* La Fig. 12 muestra una barra (eje) en vibración torsional, fija en $x=0$ y sujeta a un torque $M$ en $x=L$. La rigidez equivalente $k_{eq}$ torsional de la barra viene dada por:

\begin{align}
k_{eq}&=\frac{M}{\theta}=\frac{GJ}{L}
\label{eq12}\tag{12}
\end{align}

donde $GJ$ se llama **rigidez torsional**, siendo $G$ el módulo de corte de la barra y $J$ el momento de inercia polar (o torsional) de la barra ($m^{4}$).

<p style="text-align:center">
	<img src="torsional.png" alt="vibración torsional" width=35%/></a><br>
    <strong> Figura 12: Barra en vibración torsional considerada como resorte.</strong>
</p>

* #### Desarrollo

+ En este caso, se sigue el mismo método analítico del resorte equivalente para una barra en vibración longitudinal (axial). Como muestra la Fig. 12, la barra que tiene largo $L$, está fija en $x=0$ y sujeta a un torque $M$ en $x=L$. 

+ La figura muestra además, un desplazamiento torsional $\theta(x)$. Este parámetro, reemplaza al desplazamiento axial $u(x)$ usado en el caso de la barra en vibración longitudinal. 
+ El torque $M$ reemplaza a la fuerza $F$.
+ Una **rigidez torsional** $GJ(x)$ reemplaza a la rigidez axial $EA(x)$.

* Así, el desplazamiento torsional $\theta(x)$ va a satisfacer la ecuación diferencial:

\begin{align}
GJ(x)\,\frac{d\theta (x)}{dx}&=M(x), \quad 0<x<L
\label{eq13}\tag{13}
\end{align}

+ Ya que la barra está fija en $x=0$, entonces $\theta(x)$ satisface la condición de borde: $\theta(0)=0.$

+ Integrando la Ec. ([13](#mjx-eqn-eq13)), se obtiene el desplazamiento torsional:

\begin{align*}
\theta(x)&=\int_{0}^{x}\frac{1}{GJ(\zeta)}M(\zeta)\,d\zeta
\end{align*}

+ Si $M(x)=M$, es decir, es el mismo en cualquier punto de la barra:

\begin{align*}
\theta(x)&=M\int_{0}^{x}\frac{1}{GJ(\zeta)}\,d\zeta
\end{align*}

+ Como muestra la figura, en $x=L$, $\theta(x)$ satisface la condición de borde: $\theta(L)=\theta.$ Por lo tanto,

\begin{align*}
\theta(L)&=\theta=M\int_{0}^{L}\frac{1}{GJ(\zeta)}\,d\zeta
\end{align*}

+ Suponiendo que la rigidez equivalente, $k_{eq}$, de la barra sujeta a torsión, se puede expresar brevemente como:

\begin{align*}
k_{eq}&=\frac{M}{\theta}=\bigg[\, \int_{0}^{L}\frac{1}{GJ(\zeta)}\,d\zeta \,\bigg]^{-1}
\end{align*}

+ Si la barra es uniforme: $GJ(\zeta)=GJ={\text{constante}}$, entonces la rigidez equivalente será: 

\begin{align*}
k_{eq}&=\bigg[\,\frac{1}{GJ}\, L \,\bigg]^{-1}=\frac{GJ}{L}\quad\bigg(\frac{N\cdot m}{rad}\bigg)
\end{align*}

Por lo tanto, se obtiene el mismo resultado de la Ec. ([12](#mjx-eqn-eq12)).

+ Claramente, la analogía se extiende a barras uniformes seccionalmente en torsión

<p style="text-align:center">
	<img src="torsionseccional.png" alt="torsion seccional" width=35%/></a><br>
    <strong> Figura 13: Barras uniformes seccionalmente en vibración torsional, consideradas como resorte</strong>
</p>

+ En este caso, la rigidez equivalente será:

\begin{align}
k_{eq}&=\bigg( \frac{L_{1}}{GJ_{1}}+\frac{L_{2}}{GJ_{2}} \bigg)^{-1}
\label{eq13}\tag{13}
\end{align}
